```python
#excelファイルとして出力したいため、pandasのデータフレームを使って出力しようとした。単語出現頻度を調べるアルゴリズムとは全く関係はない。
import pandas as pd


#2次元リストに格納した単語の単語出現頻度が高い順に並べ替える際に使うバブルソートの関数の定義
def bubbleSort_REVERSE(lis):
    N=len(lis)

    for i in range(N-1):
        for j in range(N-1, i, -1):
            if lis[j][1]>lis[j-1][1]:
                lis[j], lis[j-1]=lis[j-1],lis[j]
    return lis

#テキストファイルを効率的に読み込むためにリストを作った
text_file_list=["5","5c","5s","6","6s","7","8","x","xs","xr","11","11pro"]

#excelファイルに出力するときに列名を表示させるためのもの
columns_list=["word","count"]


for h in text_file_list:
    #テキストファイルの読み込み
    with open("text_for_report/iphone{}.txt".format(h)) as f:
        data=f.readlines()
    
    #単語出現頻度を数えるときは辞書型の方がやりやすいため辞書型で数えることにする
    word_count={}
    
    #重複のない単語のリスト。辞書型のkeyのものと同じ物が入っている。自分の能力では辞書型のkeyを呼び出せず、ソートをすることができなかったため、ソートをするとき用に2次元配列の下地になるものを作っておいた。
    word_list=[]

    for line in data:
        #小文字化
        line=line.lower()

        #不要な記号を消去
        line=line.replace(",","")
        line=line.replace("\n","")
        line=line.replace("“","")
        line=line.replace("\"","")
        line=line.replace("”","")
        line=line.replace(".","")
        line=line.replace("!","")
        line=line.replace("?","")
        line=line.replace("*","")
        line=line.replace(":","")
        line=line.replace(";","")
        line=line.replace("–","")
        line=line.replace("—","")
        line=line.replace("-"," ")
        line=line.replace("®","")
        line=line.replace("™","")
        line=line.replace("("," ")
        line=line.replace(")"," ")
        line=line.replace("/"," ")
        line=line.replace("¹","")
        
        #記事中に、ios 13のように出てくる表現において、iosと13にそれぞれ分けられると面倒なのでios 13ならばios13として扱えるようにする
        line=line.replace("ios ","ios")


        #読み込んだ文章を、スペース" "を基準として1つ1つの単語として分けて単語リストして保存する。このとき、重複があるリストとなっている。
        words=list(map(str,line.split(" ")))
        
        #文章中にある全ての単語があるwordsについて操作を行う
        for i in range(len(words)):
            if words[i] not in word_count: #単語頻度集計辞書に入っていないなら
                word_count[words[i]]=1 #単語頻度集計辞書に入れ、valueは1とする
                word_list.append(words[i]) #単語頻度集計辞書のkeyと同じ要素で構成されるリストを作っておきたいので、word_listに初出単語を入れる
            else: #単語頻度集計辞書に既に入っているなら
                word_count[words[i]]+=1 #単語頻度集計辞書のvalueに1を足す
    
    #特に意味はないが単語頻度が多くなってしまった単語を削除する。
    word_list.remove("")
    word_list.remove("and")
    word_list.remove("with")
    word_list.remove("a")
    word_list.remove("in")
    word_list.remove("to")
    word_list.remove("of")
    word_list.remove("the")
    word_list.remove("is")
    word_list.remove("an")
    word_list.remove("for")
    word_list.remove("are")
    word_list.remove("on")
    word_list.remove("all")
    word_list.remove("that")
    
    #ソートの準備として、2次元配列を定義する
    word_list_for_sort=[[0 for _ in range(2)]for _ in range(len(word_list))]
    
    #2次元配列の要素を決める
    for i in range(len(word_list)):
        word_list_for_sort[i][0]=word_list[i] #2次元配列のi番目の要素の0番目は単語
        word_list_for_sort[i][1]=word_count[word_list[i]] #2次元配列のi番目の要素の1番目は単語の出現回数
    
    #2次元配列のソートをする
    word_count_sorted = bubbleSort_REVERSE(word_list_for_sort)
    
    #出現頻度上位50個を出力したい
    word_count_sorted_top50 = word_count_sorted[:50]

    #便宜上、pandasのDataframeを使って出力する
    df = pd.DataFrame(word_count_sorted_top50)
    
    #列名を作る
    df.columns = columns_list
    
    #excelに出力
    df.to_excel('excel_for_report/iphone{}.xlsx'.format(h), sheet_name='iphone {}'.format(h))
    
    print("!!!! iphone {} !!!!".format(h))
    
    for i in range(50):
        print(word_count_sorted_top50[i][0], word_count_sorted_top50[i][1])
        
    print(" ")
    print(" ")
```
